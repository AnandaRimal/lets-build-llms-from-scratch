### 4. Visualizing Perceptron on line chart

```python
#@title visualize
import matplotlib.pyplot as plt
import numpy as np

# Generate input values from -5 to 5
inputs = np.arange(-5, 6) # <-- new method numpy arange

# Calculate outputs using the defined perceptron
outputs = [myPerceptron.activate(i) for i in inputs] # <-- new syntax: python for loop

# Plotting the results
plt.figure(figsize=(8, 6))
plt.plot(inputs, outputs, marker='o', linestyle='-', color='b')
plt.title('Perceptron Output for Inputs from -5 to 5')
plt.xlabel('Input (x)')
plt.ylabel('Output (w*x + b)')
plt.grid(True)
plt.xticks(inputs) # Show all input ticks
plt.yticks(outputs) # Show all output ticks
plt.axhline(0, color='grey', linewidth=0.8) # Add x-axis line
plt.axvline(0, color='grey', linewidth=0.8) # Add y-axis line
plt.show()
```
